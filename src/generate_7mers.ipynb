```python
#@title Generate 50 random 7-mer DNA YAML files
import os
import random

output_dir = "/content/7mer_sequences"
os.makedirs(output_dir, exist_ok=True)

bases = ["A", "C", "G", "T"]

for i in range(1, 11):
    seq = "".join(random.choice(bases) for _ in range(7))
    fname = f"seq{i}_{seq}.yaml"
    yaml_text = (
        "sequences:\n"
        "  - dna:\n"
        f"      id: [{i}]\n"
        f"      sequence: {seq}\n"
    )
    with open(os.path.join(output_dir, fname), "w") as f:
        f.write(yaml_text)

print(f"Generated 50 YAML files in {output_dir}/")
print()
for fname in sorted(os.listdir(output_dir)):
    print(fname)
```
